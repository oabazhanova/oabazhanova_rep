# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] ='Daniel Radcliffe & Rupert Grint'
```

```python
df=pd.DataFrame(data.cast.str.split('|').tolist()).stack().value_counts()
pair_films = pd.DataFrame(columns=['cast pair', 'films'])
n=0
film_sum=7

for i in df.index:
    data_i = data[data.cast.str.contains(i, na=False)]['cast']
    df_i = pd.DataFrame(data_i.str.split('|').tolist()).stack().value_counts()
    for j in df_i.index:
        if j!=i and (df_i[j]>film_sum):
            pair_films.loc[n] = [i +' + '+ j, df_i[j]]
            film_sum==df_i[j]
            n+=1
         
pair_films[pair_films.films == pair_films.films.max()]
pair_films.sort_values(by='films', ascending=False)
```

ВАРИАНТ 2

```python
df=pd.DataFrame(data.cast.str.split('|').tolist()).stack().value_counts()
pair_films = pd.DataFrame(columns=['cast pair', 'films'])
n=0
film_sum=(df.max()//4)

while len(pair_films)==0:
    for i in df.index:
        data_i = data[data.cast.str.contains(i, na=False)]['cast']
        df_i = pd.DataFrame(data_i.str.split('|').tolist()).stack().value_counts()
        for j in df_i.index:
            if j!=i and (df_i[j]>film_sum):
                pair_films.loc[n] = [i +' + '+ j, df_i[j]]
                film_sum==df_i[j]
                n+=1
    film_sum-=1
            
pair_films[pair_films.films == pair_films.films.max()]
#pair_films.sort_values(by='films', ascending=False)
```
